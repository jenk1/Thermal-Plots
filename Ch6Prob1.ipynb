# Problem Statement

Two Einstein solids where the first solid contains just a single oscillator while the second one has 100 oscillators. Total number of energy units for combined system is fixed at 500. Make table multiplicity for energy from 0 to 20, graph total mutltiplicty verus energy of the first solid and then do log of total multiplicity versus energy.

```python
import numpy as np 
import matplotlib.pyplot as plt
from math import factorial 
```

Multiplicty for Einstein solid where N is number of oscillators and q is units of energy 
$\Omega(N,q) = (\frac{q + N - 1}{q})$


For more than one solid
$\sum{\Omega_{total} = (\frac{N_A + N_B + q_{total} - 1}{q_{total}})}$

```python
def mult(q, Na = 1, Nb = 100):
    x = factorial(Na + Nb + q - 1)/factorial(q)
    return(x)
```

```python
total = np.asarray([mult(i) for i in range(0,21)])
total
```

```python
# energy of the first solid
q_a = np.arange(0,21)
q_a
```

```python
# total multiplicty versus energy 
plt.plot(q_a, total)
plt.ylabel("$\Omega_{total}$", fontsize=16)
plt.xlabel("$q_A$", fontsize=16);
```

```python
# graph for ln of total mutliplicity v energy
a = np.asarray(q_a)
b = np.log(np.asarray(total))
```

```python
plt.plot(a, b)
plt.ylabel("$ln(\Omega)$", fontsize=16)
plt.xlabel("$q_a$", fontsize=16);
```
